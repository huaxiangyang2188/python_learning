```python
# 列表操作之切片，也就是区间，第一个到最后一个加一。
# 切片操作符：[]
# 切片操作符的格式：[起始:结束:步长]
# 切片的规则是：
# 1. 包含起始索引对应的元素
# 2. 不包含结束索引对应的元素
# 3. 步长默认为1，可以省略
# 4. 负数表示从右向左取值，-1表示最后一个元素
# 5. 步长可以大于1，表示间隔取值
# 6. 步长可以为空，表示取值到结束
# 7. 起始索引和结束索引都可以为空，表示相应的一端全部
# 8. 起始索引如果为空，默认为0，结束索引如果为空，默认为列表长度
# 9. 切片操作不会改变原有列表的值
# 10. 切片操作返回一个新的子列表
# 11. 切片操作不会改变原有列表的长度
# 12. 切片操作可以用于任何序列，比如字符串、元组和列表
numbers = [1, 2, 3, 4, 5]
print(numbers[1:3])  # [2, 3]
print(numbers[ :5])  # [2, 3, 4, 5]
print(numbers[-3: ])  # [ 3, 4, 5]
print(numbers[ :-3 ])  # [1,2]
# 遍历切片
for i in numbers[1:3]:
    print(i)  # 2, 3
my_pizzas = ["pepperoni", "pineapple", "anchovies"]
friend_pizzas = my_pizzas[:]
print("The first two pizzas are: ")
print(my_pizzas[0:2])
my_pizzas.append("mushrooms")
friend_pizzas.append("tomato")
print("\nMy pizzas: ")
for pizza in my_pizzas:
    print(pizza)
for pizza in friend_pizzas:
    print(pizza)
```

```python
# for循环是针对集合内每个元素的代码块，while循环是针对条件是否成立的代码块，不断运行直至条件不成立。
# 因此，for循环适用于已知的集合，while循环适用于未知的集合。
# while循环的条件设置很关键，如果条件设置错误，程序将陷入死循环。
# 在多个条件的情况下，可以通过定义变量，在循环内改变变量的值，从而控制循环。也就是标志。
# 循环内的break语句，可以跳出当前循环。
# continue语句，可以结束本次循环，不执行后续代码而直接进入下一次循环。
# 嵌套循环，可以将两个或多个循环组合在一起使用。
prompt = "\nPlease enter the ingredients for your pizza: "
prompt += "\nEnter 'quit' when you're finished."
ingredients = []
while True:
    input_ingredients = input(prompt)
    if input_ingredients == 'quit':
        break
    else:
        ingredients.append(input_ingredients)
print("Your pizza will be ready soon. Here are the ingredients you entered:")
for ingredient in ingredients:
    print("- " + ingredient)







```

```python
# 通过将列表向函数传递，提高代码的效率
def show_magicians(magicians):
    """创建一个函数，显示列表中的所有魔术师"""
    for magcian in magicians:
        print( magcian.title() + ".")
def make_great(magicians):
    """创建一个函数，将列表中的所有魔术师都变成大魔术师"""
    great_magicians = []
    while magicians:
        magician = magicians.pop()
        great_magician = magician.title() + " the great!"
        great_magicians.append(great_magician)
    """将great_magicians列表返回"""
    for great_magician in great_magicians:
        magicians.append(great_magician)
        print(great_magician)

magicians = ['Tom', 'Dick', 'Harry']
print("\n")
show_magicians(magicians)
make_great(magicians)
```

```python
from random import randint
x = randint(1,6)
class Die():
    """建立一个骰子类"""
    def __init__(self,sides=6):
        """初始化骰子类"""
        self.sides = sides

    def roll_die(self):
        """掷骰子and return a number between 1 and the number of sides"""
        return randint(1,self.sides)
# make a 6_sided die,and roll it 10 times.
d6 = Die(6)
results = []
for i in range(10):
    result = d6.roll_die()
    results.append(result)
print("The results of 10 rolls of a 6_sided die are: ")
print(results)

# make a 20_silded die,and roll it 30 times.
d20 = Die(20)
results = []
for i in range(30):
    result = d20.roll_die()
    results.append(result)
print("The results of 30 rolls of a 20_sided die are: ")
print(results)
```

```python
def get_formatted_name(city,country,population=''):
    """Generate a formatted city name"""
    if population:
        full_name = city + ' ' + country +'-'+ str(population)
    else:
        full_name = city + ' ' + country
    return full_name.title()
```
